```python
def remove_accents(s: str) -> str:

    # characters to preserve
    keep = {"č","ć","đ","š","ž","Č","Ć","Đ","Š","Ž"}
    result = []
    for ch in s:
        if ch in keep:
            result.append(ch)
        else:
            # decompose and drop all combining marks
            decomposed = unicodedata.normalize('NFD', ch)
            stripped = ''.join(c for c in decomposed if unicodedata.category(c) != 'Mn')
            result.append(stripped)
    return ''.join(result)

def transform_dict(orig: dict[str, int]) -> dict[str, int]:

    new = {}
    for word, value in orig.items():
        stripped = remove_accents(word)
        if len(word) == 1 and stripped != word and value == 0:
            new[stripped] = 1
        else:
            new[stripped] = value
    return new
```

```python
import unicodedata

def decompose_letter(ch: str):
    """
    Given a single character like 'ȁ', return a tuple:
      (base_letter, [list_of_combining_marks])
    """
    # Normalize to NFD (Normalization Form Decomposed)
    decomposed = unicodedata.normalize('NFD', ch)
    base = decomposed[0]
    combining = list(decomposed[1:])
    return base, combining

# Example:
letter = 'ȁ'
base, marks = decompose_letter(letter)
print(base)   # → 'a'
print(marks)  # → ['\u0300', '\u0300'] (two combining grave accents)

# If you want the *names* of those combining marks:
for m in marks:
    print(unicodedata.name(m))
# → COMBINING GRAVE ACCENT
# → COMBINING GRAVE ACCENT
```
